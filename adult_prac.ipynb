```python
import pandas as pd
import numpy  as np
from  sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

data = pd.read_csv("adult.csv")

data =  data.drop(['fnlwgt', 'education', 'native-country'],axis=1)
data.replace("?",np.nan,inplace=True)
data.dropna(inplace=True)
print(data.columns)

le = LabelEncoder()
data['income'] = le.fit_transform(data['income']) 

# ALWYAS FIRST SPLIT DATA THEN ENCODE It

X = data.drop('income', axis=1)
y = data['income']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)

X_train = pd.get_dummies(X_train, drop_first=True)   #drop_first=True means remove the first category column after one-hot encoding.
X_test = pd.get_dummies(X_test, drop_first=True)

# NOT ENCODE Y BECAUSE WE ENCODE FIRST BBEFORE SPLITTING

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)   #This fixes a column mismatch problem caused by one-hot encoding
X_train = X_train.astype(int)
X_test = X_test.astype(int)

# print(X_train.head())

models = {
    "Logistic Regression" : LogisticRegression(max_iter=2000),
    "KNN" : KNeighborsClassifier(),
    "decision Tree" : DecisionTreeClassifier(),
    "Random Forest" : RandomForestClassifier()
}

for name,model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(name, "Accuracy : ", accuracy_score(y_test,y_pred))
    print(name, "confusion matrix : ", confusion_matrix(y_test, y_pred))
```
